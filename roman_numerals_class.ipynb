```python
# Roman Numerals to numbers converter.

# d1 = {1: 'I', 2: 'II', 3: 'III', 4: 'IV', 5: 'V', 
#       6: 'VI', 7: 'VII', 8: 'VIII', 9: 'IX'}
# d2 = {1: 'X', 2: 'XX', 3: 'XXX', 4: 'XL', 5: 'L', 
#       6: 'LX', 7: 'LXX', 8: 'LXXX', 9: 'XC'}
# d3 = {1: 'C', 2: 'CC', 3: 'CCC', 4: 'CD', 5: 'D', 
#       6: 'DC', 7: 'DCC', 8: 'DCCC', 9: 'CM'}
# d4 = {1: 'M', 2: 'MM', 3: 'MMM', 4: 'MMMM'}
# d = [d1,d2,d3,d4]

d = [{1: 'I', 5: 'V', 10: 'X'},{1: 'X', 5: 'L', 10: 'C'},
     {1: 'C', 5: 'D', 10: 'M'},{1: 'M'}]
r = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}


class RomanNumerals:

    def to_roman(val):
        l=[]
        val = str(val)
        for index, i in enumerate(val[::-1]):
            if int(i) == 0:
                continue
            if int(i) in (1,2,3):
                l.append(int(i)*d[index][1])
            if int(i) == 4 and index != 3:
                l.append(d[index][1] + d[index][5])
            if int(i) == 5:
                l.append(d[index][5])
            if int(i) in (6,7,8):
                l.append(d[index][5] + (int(i)-5)*d[index][1])
            if int(i) == 9:
                l.append(d[index][1] + d[index][10])
            if int(i) == 4 and index == 3:
                l.append('MMMM')
        l.reverse()        
        return ''.join(l)
            
# Alternative to_roman function. It needs the above dictionaries d1,...,d4.            
#         val = str(val)
#         for k, i in enumerate(val[::-1]):
#             if int(i) == 0:
#                 continue
#             else:
#                 l.append(d[k][int(i)])                        
#         l.reverse()        
#         return ''.join(l)
        

    def from_roman(roman_num):
        if roman_num == '':
            return 0
        result = r[roman_num[-1]]
        for index, j in enumerate(roman_num[:0:-1], start = 2):
            if r[j] <= r[roman_num[-index]]:
                result = result + r[roman_num[-index]]
            else:
                result = result - r[roman_num[-index]]
        return result
```
